```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()
```

```python
wap = ['tom_1.txt', 'tom_2.txt', 'tom_3.txt', 'tom_4.txt']
text = ''

for tom in wap:
    with open('tom_1.txt', 'r', encoding='utf-8') as f:
        tmp = f.read().lower()
    tmp = tmp.replace('\n', ' ')
    tmp = ' '.join(re.sub('[^а-я ]','', tmp).split())
    text += tmp
```

```python
RUSSIAN = ' абвгдежзийклмнопрстуфхцчшщъыьэюя' #по результатам тестов буква ё в тексте не обнаружена
POS = {l: i for i, l in enumerate(RUSSIAN)}
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))

for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
    
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])

df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))

plt.figure(figsize=(20,15))
sns.heatmap(df)
plt.show()
```

**Посчитайте вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р;**

```python
s_char = 'п'
e_char = 'р'
start = POS[s_char]
end = POS[e_char]
prior = 1. / len(RUSSIAN)
for second in RUSSIAN:
    proba = prior * probabilities[start, POS[second]] 
    for third in RUSSIAN:
        proba = proba + prior * probabilities[POS[second], POS[third]] * probabilities[POS[third], end]  
        if proba > 0:
            print(''.join([s_char, second, third, e_char]), proba)
```

# Временные ряды

```python
from hmmlearn.hmm import GaussianHMM
```

```python
robberies_in_boston = pd.read_csv("../monthly-boston-armed-robberies-j.csv")
```

```python
series = robberies_in_boston['Count']
values = series.values.reshape(-1,1)
```

```python
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    print (n, model.score(values))
```

```python
model = GaussianHMM(n_components=12,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
